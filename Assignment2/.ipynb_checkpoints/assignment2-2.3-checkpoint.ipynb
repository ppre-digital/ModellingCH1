#### Import the necessary functions

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

#### Copy data and EC and RK functions from Assignment 2.1

```python
Eff0 = 0.6510   # zero loss efficiency [-]
a1=1.631        # linear coefficient for heat losses [W/(m^2 * K)]
a2=0.0096       # quadratic coefficient for heat losses [W/(m^2 * K)]
ceff=44030      # coefficient of efficeincy [J/(m^2 * K)]
cf=4180         # specific heat capacity of fluid [J/(kg * K)]
#A=1.33          # area of collector [m^2]
m=10/3600       # mass flow per unit area [kg/(s * m^2)]
Ta= 5           # ambient temperature [C]
G=700           # solar irradiance on collector's plate surface per unit area [W/m^2]
Tin=10          # inlet temperature of fluid [C]
T_out = lambda x, Tin : 2*x - Tin      # anonymous function for calculating outlet temperature of fluid [C]
```

```python
def EulerCauchy(F, t0, h, tfinal, y0):
    t = np.arange(t0,tfinal+h,h)
    y = np.ones(len(t))
    y[0] = y0
    for i in range(1,len(t)):
        s1 = F(y[i-1]) #F(t[i-1], y[i-1])
        y[i] = y[i-1] + s1*h 
    return y
```

```python
def RungeKutta(F, t0, h, tfinal, y0):
    t = np.arange(t0,tfinal+h,h)
    y= np.ones(len(t))
    y[0] = y0
    
    for i in range(1,len(t)):
        s1 = F(y[i-1])                 #F(t[i-1], y[i-1])
        s2 = F(y[i-1] + h*s1/2)        #F(t[i-1]+h/2, y[i-1]+h*s1/2)
        s3 = F(y[i-1] + h*s2/2)        #F(t[i-1]+h/2, y[i-1]+h*s2/2)
        s4 = F(y[i-1] + h*s3)          #F(t[i-1]+h, y[i-1]+h*s3)
        y[i] = y[i-1] + h*(s1 + 2*s2 + 2*s3 + s4)/6
    return y, t
```

#### Same from Assignment 2.1, however `h_tfinal` will change

```python
t0=0           # initial value for t 
hr = 5         # simulation time [h]
h_tfinal = np.array([[1, hr*60*60], [1*60, hr*60*60], [10*60, hr*60*60], 
                     [30*60, hr*60*60], 
                     [60*60, hr*60*60]])     # a matrix with the requested values for Tout calculation, 
                                             # first column is timestep size, second column is simulation time (seconds)

Tm0=10         # initial value for Tm mean fluid temperature(C)

# Function for the collector's dynamic model equation after simplification
F = lambda y: (((G*Eff0)-a1*(y-Ta)-a2*(y-Ta)**2)-(m*cf*((2*y)-(2*Tin))))/ceff
```

#### Assign the value of outlet temperature using the static model to `T_static`

```python
T_static = 45.61
```

#### Copy same loop from Assignment 2.1, add plotting for each timestep

```python
for i in range(len(h_tfinal)):     # loop for the number of rows in h_tfinal matrix
    
    h = h_tfinal[i,0]              # assign the value of timestep in h_tfinal to h
    tfinal = h_tfinal[i,1]         # assign the value of simulation time in h_tfinal to tfinal
    
    Tm_EC = EulerCauchy(F, t0, h, tfinal, Tm0)        # call the Euler-Cauchy function
    Tm_RK, t = RungeKutta(F, t0, h, tfinal, Tm0)      # call the Runge-Kutta function
    
    # calculate outlet temperature from respective mean temperatures
    Tout_EC = T_out(Tm_EC,Tin)
    Tout_RK = T_out(Tm_RK,Tin)
    
    # plot relations
    plt.figure(figsize=(10, 6))
    plt.plot(t, Tout_EC, t, Tout_RK)
    plt.axhline(y=T_static, color = 'g', linestyle='-.')    # add a horizontal line showing the outlet temperature from static model
    plt.legend(['Outlet Temperature EC', 'Outlet Temperature RK', 'Outlet Temperature - Static Model'], prop={'size': 12})
    plt.title('Solving for a time step size of {} seconds and total simulation time of {} hours' .format(h_tfinal[i,0], int(h_tfinal[i,1]/3600)));
    plt.xlabel('Simulation time [s]')
    plt.ylabel('Outlet temperature [$^\circ$C]')
    plt.grid()
    plt.show()
```

#### Conclusion

**For which purpose(s) would a steady-state approach be justified?**
<br>
A steady-state approach would be justified in this case, if the total simulation time is 3 hours (10,800 seconds) or more and we are only interested in the final temperature value, and not interested in the dynamic behaviour of the experiment.


**Which numerical integration method could be used under which conditions (e.g. time steps and total time)?**
<br>
It is clear that for small time steps and high simulation time, it would be a better approach to use the Euler Cauchy method, since it gives results very close to that of the Runge-Kutta method with less computational time. On the other hand, it is very clear, from the simulations using 30 and 60 minutes for time steps, that the Eule-Cauchy is not appropriate for usage with such high time steps and that the Runge-Kutta gives better results.
